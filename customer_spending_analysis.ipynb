```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create a fake dataset of customer spending
data = {
    "Customer": ["A", "B", "C", "D", "E", "F"],
    "Monthly_Spend": [250, 400, 150, 300, 500, 350]
}

# Put data into a DataFrame (like an Excel table)
df = pd.DataFrame(data)

# Show the table
print(df)

# Plot spending
plt.bar(df["Customer"], df["Monthly_Spend"], color="skyblue")
plt.title("Eva's Monthly Spending")
plt.xlabel("Customer")
plt.ylabel("Spend ($)")
plt.show()
```
